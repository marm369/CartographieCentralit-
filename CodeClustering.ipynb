```python
import json
from sklearn.cluster import KMeans
import numpy as np

# Chargement des donnes JSON  partir du fichier
filename = 'C:/Users/dell/monPackage/centralite.json'
with open(filename, 'r') as file:
    json_data = json.load(file)

# Nombre d'heures (supposons 24 pour votre cas)
num_hours = 24

# Structure pour stocker les rsultats de tous les clusters pour chaque heure
all_clusters = {}

# Effectuer le clustering pour chaque heure
for hour in range(num_hours):
    # Extraire les donnes pour l'heure actuelle pour toutes les rues
    X_hour = np.array([item['data'][hour] for item in json_data]).reshape(-1, 1)
    
    # Effectuer le clustering K-Means pour cette heure
    kmeans = KMeans(n_clusters=3, random_state=0).fit(X_hour)
    
    # Initialiser un dictionnaire pour regrouper les rues par cluster pour cette heure
    clusters_hour = {cluster: [] for cluster in range(3)}  # Adjustez ce nombre en fonction du nombre de clusters souhait
    
    # Grouper les rues par cluster et inclure le nom, la latitude et la longitude pour chaque rue
    for i, item in enumerate(json_data):
        cluster_label = int(kmeans.labels_[i])
        street_info = {  # Crez un dictionnaire pour chaque rue avec les informations requises
            'name': item['name'],
            'latitude': item['latitude'],
            'longitude': item['longitude']
        }
        clusters_hour[cluster_label].append(street_info)  # Ajoutez ce dictionnaire au cluster correspondant
    
    # Stocker les rsultats dans la structure all_clusters
    all_clusters[f'hour_{hour}'] = clusters_hour

# Sauvegarder tous les rsultats dans un seul fichier JSON
output_filename = 'C:/Users/dell/monPackage/all_clusters_last.json'
with open(output_filename, 'w') as outfile:
    json.dump(all_clusters, outfile, indent=4)
```
